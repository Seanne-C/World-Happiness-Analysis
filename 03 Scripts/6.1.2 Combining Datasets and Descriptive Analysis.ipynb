# Cleaning the Data

```python
wh_concat.describe()
```

```python
# Checking for mixed data types
for col in wh_concat.columns.tolist():
  weird = (wh_concat[[col]].applymap(type) != wh_concat[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (wh_concat[weird]) > 0:
    print (col)
```

```python
# Checking for missing values
wh_concat.isnull().sum()
```

```python
# checking for missing values
df_nan = wh_concat[wh_concat['Perception of Corruption'].isnull() == True]
df_nan
```
