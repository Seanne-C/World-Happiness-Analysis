## Correlations

```python
# Create a correlation matrix using pandas
wh_combined.corr()
```

```python
# Create a subset excluding the "Year" column

sub = wh_combined[['Overall Rank', 'Happiness Score', 'GDP per Capita', 'Social Support','Health', 'Freedom', 'Generosity', 'Perception of Corruption']]
```

```python
sub
```

```python
# Creating a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Creating a correlation heatmap in seaborn 
corr = sns.heatmap(sub.corr(), annot = True, ax = ax) 
```

The coefficients in the plot tell us that there are some significant relationships between variables. ## Categorical Plots

```python
# Creating a categorical plot 
# First creating a histogram

sns.histplot(wh_combined['Happiness Score'], bins = 20, kde = True)
```

```python
wh_combined.loc[wh_combined['Happiness Score'] < 4.5 , 'Happiness category'] = 'Low happiness'
```

```python
wh_combined.loc[(wh_combined['Happiness Score'] >= 4.5) & (wh_combined['Happiness Score'] < 6.5), 'Happiness category'] = 'Middle happiness'
```

```python
wh_combined.loc[wh_combined['Happiness Score'] >= 6.5, 'Happiness category'] = 'High happiness'
```

```python
wh_combined['Happiness category'].value_counts(dropna = False)
```

```python
wh_combined.columns
```

```python
# Create a categorical plot in seaborn 

sns.set(style="ticks")
g = sns.catplot(x="Country", y="Happiness Score", hue="Happiness category", data=wh_combined)
```

There are less countries with high happiness levels and more with medium to low happiness levels.
