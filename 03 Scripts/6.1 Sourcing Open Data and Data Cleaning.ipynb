```python
# Define function to swap columns
def swap_columns(wh_15_2, col1, col2):
    col_list = list(wh_15_2.columns)
    x, y = col_list.index(col1), col_list.index(col2)
    col_list[y], col_list[x] = col_list[x], col_list[y]
    wh_15_2 = wh_15_2[col_list]
    return wh_15_2
```

```python
# Swap columns
wh_15_2 = swap_columns(wh_15_2, 'Perception of Corruption', 'Generosity')
```
